**Importing Python Modules**

```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

Downloading pretrained weights for the detector and recognizer

```python
trained_data=kr.pipeline.Pipeline()
```

**Importing** **Images**

```python
ocr_images=[
           kr.tools.read(url) for url in['http://jeroen.github.io/images/testocr.png',
                                        'https://jeroen.github.io/images/utrecht2.png',
                                        'https://upload.wikimedia.org/wikipedia/commons/b/b4/EUBanana-500x112.jpg'

                                         ]
]
```

**Displaying** **the** **Image**

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[1])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[2])
```

```python
recognizing_models=trained_data.recognize(ocr_images)
```

Get Predicted Data

```python
fig, axs = plt.subplots(nrows=len(ocr_images), figsize=(15, 15))
for ax, image, predictions in zip(axs, ocr_images, recognizing_models):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
print(predictions)
```
